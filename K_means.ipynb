# K_MEANS algorithm 
- Suppose there are N points data. A matrix with dim :(d,N) and K : K < N is the predetermined number of clusters
- We need to find cluster centers m1,m2,...m_K and the label of each data points
- Element yi ∈ R 1xK , where all elements of yi are zero except the element in postion the kth postionis 1. This representation is also known as 'one-hot' encoding
- When overlapping yi vectors on top of each other , we get a label matrix  Y ∈ R NxK
- Equation of yij : yij ∈ {0, 1}, ∀i, j; sigmasum_{j=1}^{K} : yij = 1, ∀i (1)


Lost function:
- m_k ∈ R is the kth cluster center. Suppose a data point xi is assigned cluster K. Error Vector if xi is replaced by mk is (xi-mk)^2.
- We wannt this errorr vector close to zero , meaning xi is close to m_k
- Combine (1) equation : => norm2(xi-mk)^2 = sigmasum_{j=1}^{K} of yij*norm2(xi-mk)^2 (2)
- So : the error average for entire data is :#L(m,y) = sigmasum_{i=1}^{N} of (2) =>This is a function with 2 variables
Optimization :
- We need lost fuction is min value, but L(m,y) has 2 variables => We have 2 method to find min of L(m,y) :
- First : Fix M , find Y : (2) = argmin || xi - mj ||_2^2 (3). Because of (3) is the square of Euclitdean distance from point xi to mj. It can be concluded  that each point xi belongs to closed cluster centres. From there  it comes  can decude the vector label of each data point => We get min lost function
- Second : Fix Y , find M : (2) =still 2 but variable in here is M. So now to find minof LOST FUNCTION we must find argmin with derivative. => We have a new cluster centre


```python
#CODDEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEE
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
import random 
means = [[2,3],[8,3],[3,6]]
means
```

```python
cov = [[1,0], [0,1]]
cov
```

```python
N = 500
K = 3
X0 = np.random.multivariate_normal(means[0],cov,N)
X1 = np.random.multivariate_normal(means[1],cov,N)
X2 = np.random.multivariate_normal(means[2],cov,N)
X = np.concatenate((X0,X1,X2),axis = 0)
X1
```

```python
X.shape 
```

```python
original_label = np.asarray([0]*N + [1]*N + [2]*N).T
```

```python
def kmeans_init_centroids(X, k):
# randomly pick k rows of X as initial centroids
 return X[np.random.choice(X.shape[0], k, replace=False)]
```

```python
def distance(X, Y):
    return np.linalg.norm(X[:,np.newaxis,:] - Y[np.newaxis,:,:], axis=2)
def kmeans_assign_labels(X, centroids):
    # calculate pairwise distances btw data and centroids
    D = distance(X, centroids)
    # return index of the closest centroid
    return np.argmin(D, axis=1)

def has_converged(centroids, new_centroids):
    # return True if two sets of centroids are the same
    return (set([tuple(a) for a in centroids]) ==
            set([tuple(a) for a in new_centroids]))

def kmeans_update_centroids(X, labels, K):
    centroids = np.zeros((K, X.shape[1]))
    for k in range(K):
        # collect all points that are assigned to the k-th cluster
        Xk = X[labels == k, :]
        
        # update centroid coordinates
        if Xk.shape[0] > 0:
            centroids[k,:] = np.mean(Xk, axis=0)
    
    return centroids
```

```python
def kmeans(X, K):
   centroids = [kmeans_init_centroids(X, K)]
   labels = []
   it = 0 
   while True:
    labels.append(kmeans_assign_labels(X, centroids[-1]))
    new_centroids = kmeans_update_centroids(X, labels[-1], K)
    if has_converged(centroids[-1], new_centroids):
        break
    centroids.append(new_centroids)
    it += 1
   return (centroids, labels, it)
```

```python
centroids, labels, it = kmeans(X, K)
print("Centers found by our algorithm:\n", centroids[-1])
```
